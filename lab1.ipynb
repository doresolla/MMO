```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
df = pd.read_excel("price1.xlsx")
df
```

```python
%matplotlib inline
plt.scatter(df.area, df.price)
plt.xlabel("Площадь, кв.м")
plt.ylabel("Стоимость, млн.руб")
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[38]])
```

```python
reg.predict([[120]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.scatter(df.area, df.price)
plt.xlabel("Площадь, кв.м")
plt.ylabel("Стоимость, млн.руб")
plt.plot(df.area, reg.predict(df[['area']]))
```

```python
pred = pd.read_excel("prediction_price.xlsx")
pred
```

```python
p = reg.predict(pred)
```

```python
pred['predicted prices'] = p
pred
pred.to_excel('new.xlsx', index = False)
```

```python
vvp = pd.read_excel('gdprussia.xlsx')
vvp
```

```python
%matplotlib inline
plt.scatter(vvp.oilprice, vvp.gdp)
plt.xlabel("Цена за нефть")
plt.ylabel("ВВП")
```

```python
model = linear_model.LinearRegression()
model.fit(vvp[['oilprice', 'year']],vvp.gdp)
```

```python
model.predict(vvp[['oilprice', 'year']])
```

```python
%matplotlib inline
plt.scatter(vvp.oilprice, vvp.gdp)
plt.xlabel("Цена за нефть")
plt.ylabel("ВВП")
plt.plot(vvp.oilprice, model.predict(vvp[['oilprice', 'year']]))
```
